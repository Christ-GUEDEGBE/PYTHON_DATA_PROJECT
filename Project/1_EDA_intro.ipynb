```python
# Importing librairies
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset
import ast

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = pd.DataFrame(dataset['train'])

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skills : ast.literal_eval(skills) if pd.notna(skills) else skills)
```

```python
# Filter for data analysts
df_da_us = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
import seaborn as sns
# Count occurrences for each job location and select the top 10
top_locations = df_da_us['job_location'].value_counts().nlargest(10)

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the barplot with Seaborn, using a reversed blue gradient from darker to lighter at the top
sns.barplot(x=top_locations.values, y=top_locations.index, palette='Blues_r')

# Add axis labels and title
plt.xlabel("Number of Workers")
plt.ylabel("Job Location")
plt.title("Top 10 Job Locations by Number of Workers")

# Show the plot
plt.show()
```

```python
fig, ax =plt.subplots(1, 3)
dict_column = {
     'job_work_from_home':'work from home',
     'job_no_degree_mention': 'job degree req',
     'job_health_insurance': 'health insurance offered'
}


for (i, (ind, val)) in enumerate(dict_column.items()):
    df_da_us[ind].value_counts().plot(kind='pie',startangle=90, autopct='%1.1f%%',ax=ax[i])
    ax[i].set_title(val)
    ax[i].legend().set_visible(False)
    ax[i].set_ylabel('')


plt.tight_layout()
```

```python
# Count occurrences for each company name and select the top 10
top_companies = df_da_us['company_name'].value_counts().nlargest(10)

# Set the figure size
plt.figure(figsize=(12, 8))

# Create the barplot with Seaborn, using a reversed blue gradient from darker to lighter at the top
sns.barplot(x=top_companies.values, y=top_companies.index, palette='Blues_r')

# Add axis labels and title
plt.xlabel("Number of Workers")
plt.ylabel("Company Name")
plt.title("Top 10 Companies by Number of Workers")

# Show the plot
plt.show()
```
